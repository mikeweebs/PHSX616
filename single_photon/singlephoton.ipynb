# Figures for single photon experiment

```python
import matplotlib.pyplot as plt
import matplotlib
from csv import reader
import numpy as np

# Latex font
matplotlib.rcParams['mathtext.fontset'] = 'cm'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
```

## Data loading and processing

```python
f = 'singlephoton.csv'

data = []
header = []

with open(f) as f:
    reader = reader(f)
    for i,row in enumerate(reader):
        if i!=0:
            data.append(row)
        else:
            header = row

data = np.array(data)

pos = data[:,0].astype('float')
ls = data[:,1].astype('float')
ld = data[:,2].astype('float')
bs = data[:,3].astype('float')
bd = data[:,4].astype('float')
```

## Fig 1: Laser single v double normalized

```python
lsn = ls/max(ls)
ldn = ld/max(ld)

lsn_pos = pos - pos[int(np.median(np.where(lsn==max(lsn))))]-.14
ldn_pos = pos - pos[np.where(ldn==max(ldn))]

fig = plt.figure()
fig,ax = plt.subplots()
ax.plot(lsn_pos,ls,'r-',label='Single Slit')
ax.plot(ldn_pos,ld,'b-',label='Double Slit')
plt.xlabel('Position [mm]',fontsize=16)
plt.ylabel('Voltage [mV]',fontsize=16)
plt.tick_params(labelsize=16)
plt.xlim(-3.8,4.4)
plt.ylim(0,830)
plt.legend(fontsize=12)

#plt.savefig('laser.pdf')
```

## Fig 2: Bulb single v double normalized

```python
bsn = (bs/max(bs))*1.04
bdn = bd/max(bd)

bsn_pos = pos - pos[np.where(bsn==max(bsn))]+.08
bdn_pos = pos - pos[np.where(bdn==max(bdn))]-.05

fig = plt.figure()
fig,ax = plt.subplots()
ax.plot(bsn_pos,bs,'r-',label='Single Slit')
ax.plot(bdn_pos,bd,'b-',label='Double Slit')
plt.xlabel('Position [mm]',fontsize=16)
plt.ylabel('Photons/Second [Hz]',fontsize=16)
plt.tick_params(labelsize=16)
plt.xlim(-4,4)
plt.ylim(0,105)
plt.legend(fontsize=12)

#plt.savefig('bulb.pdf')
```

## Fig 2: All, normalized, wavelength normalized

```python
factor = 1.24

bsn_pos_n = (bsn_pos+.1)*factor
bdn_pos_n = (bdn_pos+.01)*factor

fig = plt.figure()
fig,ax = plt.subplots()
ax.plot(lsn_pos,lsn,'m-',label='Laser Single Slit',zorder=7)
ax.plot(ldn_pos,ldn,'r-',label='Laser Double Slit',zorder=9)
ax.plot(bsn_pos_n,bsn,'g-',label='Bulb Single Slit' ,zorder=5)
ax.plot(bdn_pos_n,bdn,'k-',label='Bulb Double Slit' ,zorder=6)
plt.xlabel('Position [mm]',fontsize=16)
plt.ylabel('Normalized Light Intensity',fontsize=16)
plt.tick_params(labelsize=16)
plt.xlim(-4.5,4.5)
plt.ylim(0,1.05)
plt.legend(fontsize=10)

#plt.savefig('laser_bulb.pdf')
```
