# Figures for single photon experiment

```python
import matplotlib.pyplot as plt
import matplotlib
import csv
import numpy as np

# Latex font
matplotlib.rcParams['mathtext.fontset'] = 'cm'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
```

## Data loading and processing

```python
f = 'singlephoton.csv'

data = []
header = []

with open(f) as f:
    reader = csv.reader(f)
    for i,row in enumerate(reader):
        if i!=0:
            data.append(row)
        else:
            header = row

data = np.array(data)

pos = data[:,0].astype('float')
ls = data[:,1].astype('float')
ld = data[:,2].astype('float')
bs = data[:,3].astype('float')
bd = data[:,4].astype('float')
```

## Fig 2: Bulb single v double normalized

```python
bsn = bs/max(bs)
bdn = bd/max(bd)

bsn_pos = pos - pos[np.where(bsn==max(bsn))]+.08
bdn_pos = pos - pos[np.where(bdn==max(bdn))]-.05

fig = plt.figure()
fig,ax = plt.subplots()
ax.plot(bsn_pos,bsn,'r-',label='Single Slit')
ax.plot(bdn_pos,bdn,'b-',label='Double Slit')
plt.xlabel('Position [mm]',fontsize=16)
plt.ylabel('Normalized Photons/Second',fontsize=16)
plt.tick_params(labelsize=16)
plt.xlim(-4,4)
plt.ylim(0,1.01)
plt.vlines(0,0,1)
plt.legend(fontsize=12)
```
